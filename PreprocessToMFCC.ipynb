```python
import os
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

def save_mel_spectrogram(audio_file, output_file):

    # Load the audio file
    y, sr = librosa.load(audio_file, sr=None)

    # Generate the Mel spectrogram
    mel_spectrogram = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128, fmax=8000)

    # Convert to log scale (dB)
    mel_spectrogram_db = librosa.power_to_db(mel_spectrogram, ref=np.max)

    # Create a new figure
    plt.figure(figsize=(10, 4))

    # Display the Mel spectrogram
    librosa.display.specshow(mel_spectrogram_db, sr=sr, x_axis='time', y_axis='mel', fmax=8000)

    # Add a color bar
    plt.colorbar(format='%+2.0f dB')

    # Set title
    plt.title('Mel Spectrogram')

    # Save the figure
    plt.savefig(output_file, bbox_inches='tight', pad_inches=0)
    plt.close()  # Close the figure to free memory

def convert_audio_to_spectrograms(audio_dir, output_dir):

    # Create output directory if it doesn't exist
    os.makedirs(output_dir, exist_ok=True)

    # Iterate over all audio files in the directory
    for filename in os.listdir(audio_dir):
        if filename.endswith('.wav') or filename.endswith('.mp3'):  # Add more formats if needed
            audio_file_path = os.path.join(audio_dir, filename)
            output_file_path = os.path.join(output_dir, f"{os.path.splitext(filename)[0]}.png")

            # Save the Mel spectrogram
            save_mel_spectrogram(audio_file_path, output_file_path)
            print(f"Saved spectrogram for {filename} as {output_file_path}")

# Example usage
audio_directory = '/content/drive/MyDrive/ICBHIpreprocessAudio/COPD'
output_directory = '/content/drive/MyDrive/MFCCICBHI/COPD'

convert_audio_to_spectrograms(audio_directory, output_directory)
```
